```python
import torch
from torchvision import datasets
```

```python
def cifar10_dataset():
    train_ds = datasets.CIFAR10(root = 'data', download=True, train = True)
    print(train_ds)
    test_ds = datasets.CIFAR10(root = 'data', download = True, train = False)
    print(test_ds)
    return
```

```python
cifar10_dataset()
```
